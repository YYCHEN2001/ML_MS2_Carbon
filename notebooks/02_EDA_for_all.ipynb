```python
import pandas as pd

data = pd.read_csv('../data/processed/dataset_all.csv')
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Statistical Ti, V, Fe, Co, Ni, Zr, Mo, Sn, W column the number of true
Ti = data['Ti'].value_counts().get(True, 0)
V = data['V'].value_counts().get(True, 0)
Fe = data['Fe'].value_counts().get(True, 0)
Co = data['Co'].value_counts().get(True, 0)
Ni = data['Ni'].value_counts().get(True, 0)
Zr = data['Zr'].value_counts().get(True, 0)
Mo = data['Mo'].value_counts().get(True, 0)
Sn = data['Sn'].value_counts().get(True, 0)
W = data['W'].value_counts().get(True, 0)

# Draw a bar chart of the number of true values of each element
# Combine the counts into a single DataFrame
elements_counts = pd.DataFrame({
    'Element': ['Ti', 'V', 'Fe', 'Co', 'Ni', 'Zr', 'Mo', 'Sn', 'W'],
    'Count': [Ti, V, Fe, Co, Ni, Zr, Mo, Sn, W]
})

# Plot the bar chart with larger font size
plt.figure(figsize=(8, 6), dpi=300)
barplot = sns.barplot(x='Element', y='Count', data=elements_counts)
plt.title('Number of True Values for Each Element', fontsize=20)
plt.xlabel('Element', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

for index, row in elements_counts.iterrows():
    barplot.text(row.name, row.Count, row.Count, color='black', ha="center", fontsize=16)

plt.show()
```

```python
sns.set_style('whitegrid')
sns.set_context("notebook", font_scale=1.5)  # 调整字体大小
plt.figure(figsize=(16, 12), dpi=300)

# 绘制 CM_type 的频率分布图，按降序排列
plt.subplot(2, 2, 1)
sns.countplot(data['CM_type'], order=data['CM_type'].value_counts().index)
plt.title('CM_type Frequency Distribution')

# 绘制 CM_morph 的频率分布图，按降序排列
plt.subplot(2, 2, 2)
sns.countplot(data['CM_morph'], order=data['CM_morph'].value_counts().index)
plt.title('CM_morph Frequency Distribution')

# 绘制 MS2_morph 的频率分布图，按降序排列
plt.subplot(2, 2, 3)
sns.countplot(data['MS2_morph'], order=data['MS2_morph'].value_counts().index)
plt.title('MS2_morph Frequency Distribution')

# 绘制 CP_morph 的频率分布图，按降序排列
plt.subplot(2, 2, 4)
sns.countplot(data['CP_morph'], order=data['CP_morph'].value_counts().index)
plt.title('CP_morph Frequency Distribution')

plt.tight_layout()
plt.show()
```

```python
# 设置图表风格和字体大小
sns.set_style('whitegrid')
sns.set_context("notebook", font_scale=1.5)

# 创建一个图形对象，并将其分成1行2列
plt.figure(figsize=(20, 8), dpi=600)

# 绘制 'SSA' 列的箱线图
plt.subplot(1, 2, 1)
sns.boxplot(data['SSA'])
plt.title('Boxplot of SSA', fontsize=32)
plt.xlabel('SSA', fontsize=28)
plt.grid(True)

# 绘制 'SSA' 列的直方图
plt.subplot(1, 2, 2)
sns.histplot(data['SSA'], kde=True, bins=30)
plt.title('Distribution of SSA', fontsize=32)
plt.xlabel('SSA', fontsize=28)
plt.ylabel('Frequency', fontsize=28)
plt.grid(True)

# 调整布局
plt.tight_layout()
plt.show()
```

```python
# 对所有的数值型特征进行相关性分析
numerical_features = data.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = data[numerical_features].corr()

# 设置图表风格和字体大小
sns.set_style('whitegrid')
sns.set_context("notebook", font_scale=1.5)

# 创建一个图形对象
plt.figure(figsize=(8, 6), dpi=300)

# 绘制相关性矩阵的热力图
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features', fontsize=20)

# 调整布局
plt.tight_layout()
plt.show()
```

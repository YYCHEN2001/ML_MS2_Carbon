```python
import pandas as pd

data = pd.read_csv('../data/processed/dataset_reduced.csv')
data
```

```python
data.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 对除了CM_type, CM_morph, MS2_morph和CP_morph之外的特征进行相关性分析
numerical_features = data.select_dtypes(include=['int64', 'float64']).columns.drop(['CM_type', 'CM_morph', 'MS2_morph', 'CP_morph'])
correlation_matrix = data[numerical_features].corr()

# 设置图表风格和字体大小
sns.set_style('whitegrid')
sns.set_context("notebook", font_scale=1)

# 创建一个图形对象
plt.figure(figsize=(8, 6), dpi=300)

# 绘制相关性矩阵的热力图
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features', fontsize=20)

# 调整布局
plt.tight_layout()
plt.show()
```

```python
# 创建一个图形对象，并设置大小
fig, axes = plt.subplots(2, 2, figsize=(16, 12), dpi=300)

# 绘制 CM_type 列的分布图
barplot = sns.countplot(ax=axes[0, 0], x='CM_type', data=data)
axes[0, 0].set_title('Distribution of CM_type')
for index, value in enumerate(data['CM_type'].value_counts().sort_index().values):
    barplot.text(index, value, value, color='black', ha="center", fontsize=16)

# 绘制 CM_morph 列的分布图
barplot = sns.countplot(ax=axes[0, 1], x='CM_morph', data=data)
axes[0, 1].set_title('Distribution of CM_morph')
for index, value in enumerate(data['CM_morph'].value_counts().sort_index().values):
    barplot.text(index, value, value, color='black', ha="center", fontsize=16)

# 绘制 MS2_morph 列的分布图
barplot = sns.countplot(ax=axes[1, 0], x='MS2_morph', data=data)
axes[1, 0].set_title('Distribution of MS2_morph')
for index, value in enumerate(data['MS2_morph'].value_counts().sort_index().values):
    barplot.text(index, value, value, color='black', ha="center", fontsize=16)

# 绘制 CP_morph 列的分布图
barplot = sns.countplot(ax=axes[1, 1], x='CP_morph', data=data)
axes[1, 1].set_title('Distribution of CP_morph')
for index, value in enumerate(data['CP_morph'].value_counts().sort_index().values):
    barplot.text(index, value, value, color='black', ha="center", fontsize=16)

# 调整布局
plt.tight_layout()
plt.show()
```

```python
# 绘制float64类型的特征的分布图
fig, axes = plt.subplots(3, 3, figsize=(16, 12), dpi=300)

# 绘制 EN 列的分布图
sns.histplot(ax=axes[0, 0], x='EN', data=data, kde=True)
axes[0, 0].set_title('Distribution of EN')

# 绘制 CR 列的分布图
sns.histplot(ax=axes[0, 1], x='CR', data=data, kde=True)
axes[0, 1].set_title('Distribution of CR')

# 绘制 SSA 列的分布图
sns.histplot(ax=axes[0, 2], x='SSA', data=data, kde=True)
axes[0, 2].set_title('Distribution of SSA')

# 绘制 C 列的分布图
sns.histplot(ax=axes[1, 0], x='C', data=data, kde=True)
axes[1, 0].set_title('Distribution of C')

# 绘制 PW 列的分布图
sns.histplot(ax=axes[1, 1], x='PW', data=data, kde=True)
axes[1, 0].set_title('Distribution of PW')

# 绘制 CD 列的分布图
sns.histplot(ax=axes[1, 2], x='CD', data=data, kde=True)
axes[1, 2].set_title('Distribution of CD')

# 绘制 Cs 列的分布图
sns.histplot(ax=axes[2, 0], x='Cs', data=data, kde=True)
axes[2, 0].set_title('Distribution of Cs')

# 调整布局
plt.tight_layout()
plt.show()
```
